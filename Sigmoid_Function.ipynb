```python
# A sigmoid function serves as an activation function in our neural network training.
# So to get values between 0 and 1 we use the sigmoid function. 
# The sigmoid function converts our output values between 0 and 1.
# Import required Library
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating input with the scale of -10 to 10 
input = np.linspace(-10,10,100)
```

```python
# Generating the sigmoid function 
def sigmoid(X):
    val = 1/(1+np.exp(-X))
    return val
```

```python
# Printing output values
output = sigmoid(input)
```

```python
# Plotting the sigmoid function graph
plt.plot(input,output)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("sigmoid Function")
```

```python
plt.plot(input,output,c="b")
output = sigmoid(0.5*input)
plt.plot(input,output,c="r")
output = sigmoid(2.5*input)
plt.plot(input,output,c="g")
```
